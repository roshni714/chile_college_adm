```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from tqdm import tqdm
from utils import create_translation_dic, get_enrolled_ids_year, get_low_socio_ids_year, get_on_time_grad_ids_year
```

```python
years = list(range(2007, 2016))
prop = []
threshold = 5
for year in tqdm(years):
    enrolled_ids = get_enrolled_ids_year(year)
    low_socio_ids = get_low_socio_ids_year(year, threshold)
    n_intersection= len(enrolled_ids.intersection(low_socio_ids))
    n_enrolled = len(enrolled_ids)
    prop.append(n_intersection/n_enrolled)
```

```python
plt.plot(years, prop)
plt.xlabel("Admission Year")
plt.title("Proportion of Enrolled Students with Low Socioeconomic Status \n Family Income Threshold = 5")
plt.ylabel("Proportion")
plt.xlabel("Admission Year")
plt.tight_layout()
plt.savefig("figs/low_socio_enrolled.pdf")
```

```python
years = list(range(2007, 2016))
prop_on_time_grad = []

for year in tqdm(years):
    enrolled_ids = get_enrolled_ids_year(year)
    on_time_grad_ids = get_on_time_grad_ids_year(year)
    n_intersection= len(enrolled_ids.intersection(on_time_grad_ids))
    n_enrolled = len(enrolled_ids)
    prop_on_time_grad.append(n_intersection/n_enrolled)
    

```

```python
plt.plot(years[:len(prop_on_time_grad)], prop_on_time_grad)
plt.xlabel("Admission Year")
plt.title("Proportion of On-Time Graduates with Low Socioeconomic Status \n Family Income Threshold = 5")
plt.ylabel("Proportion")
plt.xlabel("Admission Year")
plt.tight_layout()
plt.savefig("figs/low_socio_grad.pdf")
```
